#### Equation 1.4 [5 points]

\begin{equation*}
    \frac{dy}{dx} + \frac{4y}{x} = x^{4}
\end{equation*}

This equation can be represented in the form of 

\begin{equation*}
    \frac{dy}{dx} + Py = Q                                   
\end{equation*}

where P and Q are completely dependent on x and Integrating Factor is =

\begin{equation*}
    I = e^{\int Pdx}
\end{equation*}

Guessing the value of Integrating Factor and multiplying on both sides we get
\begin{equation*}
    I\frac{dy}{dx} + IPy = IQ
\end{equation*}

Integrating both sides, 
\begin{equation*}
    \int (I\frac{dy}{dx} + IPy)dx = \int (IQ)dx
\end{equation*}

by property of Chain Rule, the Left Hand Side condenses into 

\begin{equation*}
      Iy = \int (IQ)dx                  
\end{equation*}

Calculation of Integrating Factor for our equation- 

value of P = $\frac{4}{x}$

therefore, the value of Integrating Factor is = 
\begin{equation*}
    I = e^{\int \frac{4}{x}dx}
\end{equation*}

\begin{equation*}
    I = e^{4\ln x}
\end{equation*}    

\begin{equation*}
    I = x^{4}
\end{equation*}

Proceeding with the aforementioned steps and pluggin values in the final step we get, 
\begin{equation*}
    x^{4}y = \int x^{4}x^{4}dx
\end{equation*} 

Solving the integral, 

\begin{equation*}

\end{equation*}    

rearranging the terms a bit,

## Numerical methods for solving differential equations [30 points]

Let's suppose that we have a scenario in which a robot is supposed to replace a lamp in a factory, a task which involves pressing a switch to turn the lamp off, waiting until the lamp cools down to $30^{\circ}C$, and then unscrewing it. The robot is able to measure the temperature of the lamp as long as it is turned on, but once the lamp is turned off, the temperature cannot be measured.

According to [Newton's law of cooling](http://www.ugrad.math.ubc.ca/coursedoc/math100/notes/diffeqs/cool.html), the rate of change of temperature of an object is directly proportional to the difference between the temperature of the object and the ambient temperature.

We can write this in the form of a differential equation as

\begin{equation*}
    \frac{dT}{dt} = -K(T - T_A)
\end{equation*}

where $T$ is the temperature of the object as a function of time, $T_A$ is the ambient temperature (which we assume remains constant), and $K$ is a constant that depends on the properties of the object.

In the factory, the ambient temperature is known to be $T_A = 20^{\circ}C$ and, in addition, the robot knows that for this lamp, $K = 0.05$. Our robot also queries the temperature of the lamp just before turning the lamp off and measures $T(0) = 150^{\circ}C$.

The robot needs to calculate how long it needs to wait until the lamp has cooled down sufficiently, i.e. find the time $t$ at which the temperature is just below $30^{\circ}C$.

### Analytical solution [10 points]

Your first task is to use *sympy* to find the general solution to the above problem.

```python
# YOUR CODE HERE
K,Ta,t = sp.symbols('K Ta t')
T = sp.Function('T')(t)
T_prime = sp.Derivative(T,t)
eq = T_prime + K*T- K*Ta

# please assign soln to the solution of the differential equation
eq = eq.subs([(K, 0.05),(Ta, 20)])
soln = sp.dsolve(eq, ics ={T.subs(t,0): 150})
print(soln)
```

Now that you have the general solution, use sympy to find the particular solution using the given initial conditions and then solve for the time $t$ at which the desired temperature is reached.

```python
# YOUR CODE HERE
# please assign t_30 to the resulting time
t_30 = sp.solve(soln.subs(T,30),t)[0]
print(t_30)
```

### Euler's method [20 points]

In the lab class, we already looked at how Euler's method can be used for estimating the solution to an ODE numerically. In this exercise, you will simply adapt what we did in the lab class and solve our current problem, namely that of finding the time at which the temperature of the lamp reaches $30^{\circ}C$.

#### Euler's method: Implementation [10 points]

Your first task in this exercise is to implement Euler's method for finding a solution to our problem.

```python
# initial values; don't change these
K = 0.05
T_A = 20.0
init_temp = 150.0
t_0 = 0.
stop_temp = 30.0
```

```python
import math 
def T_prime(t: float, T: float, T_A: float, K: float) -> float:
    '''Returns y' as a function of t, y, and the
    given constants T_A and K
    '''
    T_prime = -K*130*math.exp(-K*t)
    #T_prime = - K*(T-T_A)
    return T_prime

def euler(T_prime: typing.Callable,
          ics: np.array, T_A: float,
          K: float, stop_temp: float,
          h: float) -> tuple:
    '''Uses Euler's method for finding the time
    and the iteration count at which
    the temperature of the lamp reaches stop_temp.

    Keyword arguments:
    f_prime -- a callback function for calculating
               the derivative of T
    ics -- initial time and temperature
    T_A -- ambient temperature
    K -- a constant factor
    stop_temp -- desired temperature
    h -- step size
    '''
    t_current = ics[0]
    T_current = ics[1]
    ts = [T_current]
    #count = 0
    while T_current > stop_temp :
        T_current = T_current + h * T_prime(t_current, T_current, T_A, K)
        t_current += h
        ts.append(T_current)
        #count += 1
    return (ts)
```

#### Euler's method: Observations [10 points]

Now that you've implemented the method:

1. call it with different values of $h$ in the range $[0.01, 1]$
2. find the estimation error and the number of iterations required to find a solution for each value of $h$, and
3. plot the errors and the iteration counts to show how the estimation error and the number of iterations are affected by the value of $h$.

```python
ics = [t_0, init_temp]
errors = list()
iterations = list()

# YOUR CODE HERE
#t = list()
h = np.linspace(0.01, 1, 10)
count=0

for step in h:
    count += 1
    ics = [t_0, init_temp]
    
    t_euler_small_step = euler(T_prime, ics, T_A, K, stop_temp, step)
    #actual n_new size is length of data - 1
    
    n_new = len(t_euler_small_step)-1
    desired_t = n_new*step
    print ('Graph: ',count)
    print ('desired time to attain the temperature of 30 degree centigrades:', desired_t)
    print ('The number of iteration required to attain temperature of 30 degrees with step {} is: {} '.format(step,n_new))
    
    #calculating actual function
    
    t = np.linspace(0, n_new*step, n_new+1)
    y = T_A + ((150-T_A) * np.exp(-0.05*t))

    plt.figure(1, figsize=(8,5))
    plt.title('Eulers plot v/s actual function with step size: {}'.format(step))
    plt.xlabel('Time')
    plt.ylabel('Temperature')
    plt.plot(np.linspace(0, n_new*step, n_new+1), y, color='b', label = 'Function Value')
    #plotting for number of n
    plt.plot(np.linspace(0, n_new*step, n_new+1), t_euler_small_step, color='r', label = 'Euler calculation')
    plt.legend()
    
#   now calculating and plotting the estimation error graph
#   calculated between actual function and Eulers method
    plt.figure(2, figsize=(8,5))
    plt.title('Error graph')
    plt.xlabel('Time')
    plt.ylabel('Error')
    error_estimation = y-t_euler_small_step
    plt.plot(np.linspace(0,n_new*step,n_new+1), error_estimation, label = 'Estimation error ')
    
    plt.legend()
    plt.show()
```
